below the three colms are in object datatype but inside datas are int so we are converting the datatype of the columns.

````python
text_columns=['packed_cell_volume','white_blood_cell_count','red_blood_cell_count']
for i in text_columns:
  print(f"{i} : {df_data[i].dtype}")
````

````python
#we are converting only the three columns mentioned above to numerical datatype
def convert_text_to_numeric(df_data,column):
  df_data[column]=pd.to_numeric(df_data[column],errors='coerce')

#function calling
for column in text_columns:
  convert_text_to_numeric(df_data,column)
  print(f"{column}: {df_data[column].dtype}")
````

````python
#here we are finding null value in each columns and sort the number of nulls in descending order.
missing = df_data.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
````

````python
#mean is average value of all data in the column so it can only be used in int and float datatype column
#mode is highest number of occurence of a value in the column, so mode can contain many values as two or more value can be same number of times, so only we use mode()[0] to select first number with highest number of occurence
#mean is used only in int and float and mode is used in object type
def mean_value_imputation(df_data,column):
  mean_value = df_data[column].mean()
  df_data[column].fillna(value=mean_value, inplace=True)

def mode_value_imputation(df_data,column):
  mode_value = df_data[column].mode()[0]
  df_data[column]=df_data[column].fillna(mode_value)
````

````python
#here we separate columns from df_data which are numerical datatype like int, float and we fill null value with mean
num_cols = [col for col in df_data.columns if df_data[col].dtype!='object']
for col_name in num_cols:
  mean_value_imputation(df_data, col_name)
````

````python
#here we separate columns from df_data which are categorical values or object type data we fill null value with mode
obj_cols = [col for col in df_data.columns if df_data[col].dtype=='object']
for col_name in obj_cols:
  mode_value_imputation(df_data, col_name)
````

````python
#this to check whether we successfully filled all null data in the columns. and the result is empty list so successfull
missing = df_data.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
````

````python
#these three columns are one of the object type columns so it needs special attention, in output we can see that whitespace,uppercase etc makes it different. so it makes the ml wrong.
print(f"diabetes_mellitus :- {df_data['diabetes_mellitus'].unique()}")
print(f"coronary_artery_disease :- {df_data['diabetes_mellitus'].unique()}")
print(f"classification :- {df_data['classification'].unique()}")
````

````python
#we are removing uppercase, whitespace in the yes, no
df_data['diabetes_mellitus'] = df_data['diabetes_mellitus'].replace(to_replace={" yes":"yes","\tyes":"yes","\tno":"no"})
df_data['coronary_artery_disease'] = df_data['coronary_artery_disease'].replace(to_replace={" yes":"yes","\tno":"no","\tyes":"yes"})
df_data['classification'] = df_data['classification'].replace(to_replace={"ckd\t":"ckd"})
````

````python
print(f"diabetes_mellitus :- {df_data['diabetes_mellitus'].unique()}")
print(f"coronary_artery_disease :- {df_data['diabetes_mellitus'].unique()}")
print(f"classification :- {df_data['classification'].unique()}")
````

````python
#feature encoding: we changing categorical values like yes or no is changed to 0 or 1 as ml can ony works on numerical values.
#as we have only two values like yes or no we use mapping .
df_data['classification'] = df_data['classification'].map({'ckd': 1, 'notckd': 0})
df_data['red_blood_cells'] = df_data['red_blood_cells'].map({'normal':1,'abnormal':0})
df_data['pus_cell'] = df_data['pus_cell'].map({'normal':1, 'abnormal':0})
df_data['pus_cell_clumps'] = df_data['pus_cell_clumps'].map({'present':1,'notpresent':0})
df_data['bacteria'] = df_data['bacteria'].map({'present':1,'notpresent':0})
df_data['hypertension'] = df_data['hypertension'].map({'yes':1,'no':0})
df_data['diabetes_mellitus'] = df_data['diabetes_mellitus'].map({'yes':1,'no':0})
df_data['coronary_artery_disease'] = df_data['coronary_artery_disease'].map({'yes':1,'no':0})
df_data['appetite'] = df_data['appetite'].map({'good':1,'poor':0})
df_data['pedal_edema'] = df_data['pedal_edema'].map({'yes':1,'no':0})
df_data['anemia'] = df_data['anemia'].map({'yes':1,'no':0})

````

````python
from sklearn.model_selection import train_test_split
````

````python
x = df_data.drop("classification",axis=1)
y = df_data["classification"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 25)
print(f"'x' shape : {x_train.shape}")
print(f"'xtest' shape : {x_test.shape}")
````

````python
from sklearn.tree import DecisionTreeClassifier
dct = DecisionTreeClassifier()
dct.fit(x_train, y_train)
````

````python
y_pred_dct = dct.predict(x_test)
y_pred_dct
````

````python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
````

````python
models = []
models.append(('naive bayes',GaussianNB()))
models.append(('KNN', KNeighborsClassifier(n_neighbors=8)))
models.append(('RandomForestClassifier', RandomForestClassifier()))
models.append(('decisionTreeClassifier', DecisionTreeClassifier()))
models.append(('svm', SVC(kernel='linear')))
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
````

````python
for name, model in models:
  print(name, model)
  print()
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  print(confusion_matrix(y_test,y_pred))
  print('\n')
  print(accuracy_score(y_test,y_pred))
  print('\n')
  print(precision_score(y_test,y_pred))
  print('\n')
  print(recall_score(y_test,y_pred))
  print('\n')
  print(f1_score(y_test,y_pred))
  print('\n')
````
