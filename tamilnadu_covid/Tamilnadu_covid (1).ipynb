# `iloc` and `loc` are two methods in pandas for selecting data from a DataFrame:

1. **`iloc`**: Uses integer-based indexing to select rows and columns by their position in the DataFrame, similar to array indexing in Python.
   
2. **`loc`**: Uses label-based indexing to select rows and columns by their labels (index or column names), providing more flexibility when dealing with labeled data.

3. **Usage**: `iloc[row_index, column_index]` selects data by integer positions, while `loc[row_label, column_label]` selects data by labels.

4. **Inclusivity**: `iloc` excludes the end index in slicing (`[start:end]`), while `loc` includes both start and end labels (`[start:end]`).

5. **Examples**: `df.iloc[0:3, 1:3]` selects rows 0 to 2 and columns 1 to 2, while `df.loc['A':'C', 'B':'D']` selects rows labeled 'A' to 'C' and columns labeled 'B' to 'D'.

- **`plotly.offline`**: This module in Plotly is used to generate and display Plotly graphs offline, meaning you don't need an active internet connection to create or view the plots. It allows you to save the generated plots as standalone HTML files that include all necessary JavaScript and CSS.

- **`plotly.graph_objs` (`go`)**: This module provides the core functionality to define and configure the visual aspects of Plotly graphs. It allows you to create graph objects such as traces, which specify the data, layout, and styling options for your plot.

In summary, `plotly.offline` is used to enable offline plotting capabilities, while `plotly.graph_objs` (`go`) is used to define and configure the content and appearance of the Plotly graphs themselves.

 If you do not use one of the components (`plotly.offline` or `plotly.graph_objs`) in your Plotly workflow, specific issues may arise:

1. **Not Using `plotly.offline`**:
   - **Online Dependency**: You'll need an active internet connection to generate and display Plotly graphs, as they rely on loading JavaScript and CSS resources from Plotly's servers.
   - **Limited Offline Capability**: You won't be able to save Plotly graphs as standalone HTML files that include all necessary resources for complete offline viewing or sharing.

2. **Not Using `plotly.graph_objs` (`go`)**:
   - **Lack of Graph Object Configuration**: You won't be able to define and configure the content and appearance of Plotly graphs (such as data, layout, and styling options).
   - **Inability to Create Graphs**: Without `plotly.graph_objs`, you cannot create or customize Plotly graph objects (like traces), which are essential for specifying what data to plot and how to visualize it.

Sure, here are the concise differences between a Pandas `Series` and a NumPy array:

1. **Indexing**:
   - **Series**: Has labeled indices for each element.
   - **Array**: Uses integer indices starting from 0.

2. **Data Types**:
   - **Series**: Can hold heterogeneous data types.
   - **Array**: Requires homogeneous data types.

3. **Operations**:
   - **Series**: Supports Pandas-specific operations like label-based indexing and data alignment.
   - **Array**: Optimized for numerical operations and supports vectorized computations.

4. **Integration**:
   - **Series**: Part of Pandas for data manipulation and analysis.
   - **Array**: Core data structure in NumPy for numerical computing and scientific computing.

5. **Use Cases**:
   - **Series**: Ideal for labeled data and data analysis tasks.
   - **Array**: Used for numerical computations, matrix operations, and interfacing with other numerical libraries.

```python
def create_dataset(dataset, look_back=1):
  dataX,dataY = [],[]
  for i in range(len(dataset)-look_back-1):
    a = dataset[i:(i+look_back),0]
    dataX.append(a)
    dataY.append(dataset[i+look_back, 0])
  return np.array(dataX), np.array(dataY)
```

```python
look_back = 2
trainX, trainY = create_dataset(train, look_back = look_back)
testX, testY = create_dataset(test, look_back = look_back)
```
