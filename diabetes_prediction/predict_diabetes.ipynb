````python
#train and test
from sklearn.model_selection import train_test_split
````

````python
#here train_test_split needs data and label(output of the tabel) separetly.
#eg. age, gender, weight are comes under data and diabetes is present or not, which is showed in output column as 0 or 1 is the label as this is a supervised learning.
x = df.iloc[:, df.columns!='Outcome'] # this is for including all cols excluding Outcome column.
y = df.iloc[:, df.columns=='Outcome'] # this is to only include Output cols.
````

````python
print(x)
````

````python
print(y)
````

````python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2)
````

````python
#see the random index which will be same as xtrain and ytrain.
xtrain.head()
````

````python
ytrain.head()
````

````python
#xtrain is 80%
xtrain.shape
````

````python
from sklearn.ensemble import RandomForestClassifier
````

````python
model=RandomForestClassifier()
````

````python
model.fit(xtrain,ytrain.values.ravel())
````

````python
#this the value predicted from the xtrain. xtest given below is outcome column.
predict_output=model.predict(xtest)
print(predict_output)
````

````python
#to compare the prediced output with actual output we use accuarcy_score
from sklearn.metrics import accuracy_score
acc = accuracy_score(predict_output,ytest)
print("accuracy score: ",acc)
````

### Basic Usage

# **ensemble**
`sklearn.ensemble` in scikit-learn provides methods for ensemble learning, combining predictions from multiple models to improve accuracy and robustness:

1. **Bagging**: Uses multiple base models (e.g., decision trees) trained on random subsets of data for aggregation.
2. **Boosting**: Builds models sequentially, each correcting errors of its predecessor, like AdaBoost and Gradient Boosting.
3. **Random Forest**: Constructs multiple decision trees during training and averages their predictions for improved accuracy and overfitting control.
4. **Extra Trees**: Similar to Random Forest but with random feature selection to further diversify trees.
5. **Voting**: Combines predictions from multiple models by majority voting (for classification) or averaging (for regression).
6. **Stacking**: Trains a meta-model to combine predictions from multiple base models.
  
These methods are effective for various machine learning tasks, enhancing model performance through ensemble techniques.
